```python
import pandas as pd
```

```python
df = pd.read_csv("../data/ipl_2026_deliveries.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.groupby("striker")["runs_of_bat"].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
top_batsmen = df.groupby("striker")["runs_of_bat"].sum().sort_values(ascending=False).head(10)
```

```python
top_batsmen.plot(kind="bar")

plt.title("Top 10 Run Scorers - IPL 2026")
plt.xlabel("Batsmen")
plt.ylabel("Runs")

plt.show()
```

```python
sixes = df[df["runs_of_bat"] == 6]
```

```python
sixes.head()
```

```python
top_six_hitters = sixes.groupby("striker").size().sort_values(ascending=False).head(10)
```

```python
top_six_hitters
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

top_six_hitters.plot(kind="bar")

plt.title("Top 10 Six Hitters - IPL 2026")
plt.xlabel("Batsmen")
plt.ylabel("Number of Sixes")

plt.xticks(rotation=45)

plt.show()
```

```python
top_wicket_takers = df.groupby("bowler")["player_dismissed"].count().sort_values(ascending=False).head(10)

top_wicket_takers
```

```python
df["wicket_type"].value_counts()
```

```python
bowler_wickets = df[
    df["wicket_type"].isin([
        "bowled",
        "caught",
        "lbw",
        "stumped",
        "caught and bowled",
        "hit wicket"
    ])
]

top_wickets = (
    bowler_wickets.groupby("bowler")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_wickets
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

top_wickets.plot(kind="bar")

plt.title("Top 10 Wicket Takers - IPL 2026")
plt.xlabel("Bowlers")
plt.ylabel("Wickets")

plt.xticks(rotation=45)

plt.show()
```

```python
team_runs = (
    df.groupby("batting_team")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
)

team_runs
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

team_runs.plot(kind="bar")

plt.title("Team-wise Total Runs - IPL 2026")
plt.xlabel("Teams")
plt.ylabel("Runs")

plt.xticks(rotation=45)

plt.show()
```

```python
highest_scores = (
    df.groupby(["match_id", "striker"])["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

highest_scores
```

```python
highest_scores_by_player = highest_scores.reset_index()

plt.figure(figsize=(10,6))

plt.bar(highest_scores_by_player["striker"],
        highest_scores_by_player["runs_of_bat"])

plt.title("Top 10 Highest Individual Scores - IPL 2026")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.xticks(rotation=45)

plt.show()
```

```python
fours = df[df["runs_of_bat"] == 4]

fours.head()
```

```python
top_four_hitters = (
    fours.groupby("striker")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_four_hitters
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

top_four_hitters.plot(kind="bar")

plt.title("Top 10 Boundary Hitters (4s) - IPL 2026")
plt.xlabel("Batsmen")
plt.ylabel("Number of Fours")

plt.xticks(rotation=45)

plt.show()
```

```python
dot_balls = df[df["runs_of_bat"] == 0]

dot_balls.head()
```

```python
top_dot_bowlers = (
    dot_balls.groupby("bowler")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_dot_bowlers
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

top_dot_bowlers.plot(kind="bar")

plt.title("Top 10 Bowlers by Dot Balls - IPL 2026")
plt.xlabel("Bowlers")
plt.ylabel("Dot Balls")

plt.xticks(rotation=45)

plt.show()
```

```python
bowler_wickets = df[
    df["wicket_type"].isin([
        "bowled",
        "caught",
        "lbw",
        "stumped",
        "caught and bowled",
        "hit wicket"
    ])
]

bowler_wickets.head()
```

```python
best_figures = (
    bowler_wickets.groupby(["match_id", "bowler"])
    .size()
    .sort_values(ascending=False)
    .head(10)
)

best_figures
```

```python
best_figures_df = best_figures.reset_index()
best_figures_df.columns = ["Match ID", "Bowler", "Wickets"]

best_figures_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.bar(best_figures_df["Bowler"], best_figures_df["Wickets"])

plt.title("Top 10 Best Bowling Figures - IPL 2026")
plt.xlabel("Bowler")
plt.ylabel("Wickets")

plt.xticks(rotation=45)

plt.show()
```

```python
highest_team_scores = (
    df.groupby(["match_id", "innings", "batting_team"])["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

highest_team_scores
```

```python
highest_team_scores_df = highest_team_scores.reset_index()
highest_team_scores_df.columns = ["Match ID", "Innings", "Team", "Runs"]

highest_team_scores_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.bar(highest_team_scores_df["Team"],
        highest_team_scores_df["Runs"])

plt.title("Top 10 Highest Team Scores - IPL 2026")
plt.xlabel("Teams")
plt.ylabel("Runs")

plt.xticks(rotation=45)

plt.show()
```

```python
venue_matches = (
    df.groupby("venue")["match_id"]
    .nunique()
    .sort_values(ascending=False)
)

venue_matches
```

```python
plt.figure(figsize=(12,6))

venue_matches.plot(kind="bar")

plt.title("Matches Played at Each Venue - IPL 2026")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")

plt.xticks(rotation=90)

plt.show()
```

```python
venue_runs = (
    df.groupby("venue")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
)

venue_runs
```

```python
plt.figure(figsize=(12,6))

venue_runs.plot(kind="bar")

plt.title("Total Runs by Venue - IPL 2026")
plt.xlabel("Venue")
plt.ylabel("Runs")

plt.xticks(rotation=90)

plt.show()
```

```python
venue_sixes = (
    df[df["runs_of_bat"] == 6]
    .groupby("venue")
    .size()
    .sort_values(ascending=False)
)

venue_sixes
```
