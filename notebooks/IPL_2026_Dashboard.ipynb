```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/ipl_2026_deliveries.csv")
```

```python
df.head()
```

```python
top_batsmen = (
    df.groupby("striker")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_batsmen_df = top_batsmen.reset_index()
top_batsmen_df.columns = ["Batsman", "Runs"]
```

```python
fig1 = px.bar(
    top_batsmen_df,
    x="Batsman",
    y="Runs",
    title="Top 10 Run Scorers",
    text="Runs",
    color="Runs",
    color_continuous_scale="Blues"
)

fig1.update_traces(textposition="outside")
```

```python
top_six_hitters = (
    df[df["runs_of_bat"] == 6]
    .groupby("striker")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_six_hitters_df = top_six_hitters.reset_index()
top_six_hitters_df.columns = ["Batsman", "Sixes"]
```

```python
fig2 = px.bar(
    top_six_hitters_df,
    x="Batsman",
    y="Sixes",
    title="Top 10 Six Hitters",
    text="Sixes",
    color="Sixes",
    color_continuous_scale="Reds"
)

fig2.update_traces(textposition="outside")
```

```python
bowler_wickets = df[
    df["wicket_type"].isin([
        "bowled",
        "caught",
        "lbw",
        "stumped",
        "caught and bowled",
        "hit wicket"
    ])
]

top_wickets = (
    bowler_wickets.groupby("bowler")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_wickets_df = top_wickets.reset_index()
top_wickets_df.columns = ["Bowler", "Wickets"]
```

```python
fig3 = px.bar(
    top_wickets_df,
    x="Bowler",
    y="Wickets",
    title="Top 10 Wicket Takers",
    text="Wickets",
    color="Wickets",
    color_continuous_scale="Greens"
)

fig3.update_traces(textposition="outside")
```

```python
team_runs = (
    df.groupby("batting_team")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
)

team_runs_df = team_runs.reset_index()
team_runs_df.columns = ["Team", "Runs"]
```

```python
fig4 = px.bar(
    team_runs_df,
    x="Team",
    y="Runs",
    title="Team-wise Total Runs",
    text="Runs",
    color="Runs",
    color_continuous_scale="Purples"
)

fig4.update_traces(textposition="outside")
```

```python
highest_scores = (
    df.groupby(["match_id", "striker"])["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

highest_scores_df = highest_scores.reset_index()
highest_scores_df.columns = ["Match ID", "Batsman", "Runs"]
```

```python
fig5 = px.bar(
    highest_scores_df,
    x="Batsman",
    y="Runs",
    title="Top 10 Highest Individual Scores",
    text="Runs",
    color="Runs",
    color_continuous_scale="Viridis",
    hover_data=["Match ID"]
)

fig5.update_traces(textposition="outside")
```

```python
top_fours = (
    df[df["runs_of_bat"] == 4]
    .groupby("striker")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_fours_df = top_fours.reset_index()
top_fours_df.columns = ["Batsman", "Fours"]
```

```python
fig6 = px.bar(
    top_fours_df,
    x="Batsman",
    y="Fours",
    title="Top 10 Players with Most Fours",
    text="Fours",
    color="Fours",
    color_continuous_scale="Blues"
)

fig6.update_traces(textposition="outside")
```

```python
top_dot_balls = (
    df[df["runs_of_bat"] == 0]
    .groupby("bowler")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_dot_balls_df = top_dot_balls.reset_index()
top_dot_balls_df.columns = ["Bowler", "Dot Balls"]
```

```python
fig7 = px.bar(
    top_dot_balls_df,
    x="Bowler",
    y="Dot Balls",
    title="Top 10 Bowlers with Most Dot Balls",
    text="Dot Balls",
    color="Dot Balls",
    color_continuous_scale="Oranges"
)

fig7.update_traces(textposition="outside")
```

```python
best_bowling = (
    bowler_wickets.groupby("bowler")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

best_bowling_df = best_bowling.reset_index()
best_bowling_df.columns = ["Bowler", "Wickets"]
```

```python
fig8 = px.bar(
    best_bowling_df,
    x="Bowler",
    y="Wickets",
    title="Best Bowling Figures",
    text="Wickets",
    color="Wickets",
    color_continuous_scale="Greens"
)

fig8.update_traces(textposition="outside")
```

```python
highest_team_scores = (
    df.groupby(["match_id", "innings", "batting_team"])["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

highest_team_scores_df = highest_team_scores.reset_index()
highest_team_scores_df.columns = ["Match ID", "Innings", "Team", "Runs"]
```

```python
fig9 = px.bar(
    highest_team_scores_df,
    x="Team",
    y="Runs",
    title="Top 10 Highest Team Scores",
    text="Runs",
    color="Runs",
    color_continuous_scale="Plasma",
    hover_data=["Match ID", "Innings"]
)

fig9.update_traces(textposition="outside")
```

```python
venue_analysis = (
    df.groupby("venue")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
)

venue_analysis_df = venue_analysis.reset_index()
venue_analysis_df.columns = ["Venue", "Runs"]
```

```python
fig10 = px.bar(
    venue_analysis_df,
    x="Venue",
    y="Runs",
    title="Venue-wise Total Runs",
    text="Runs",
    color="Runs",
    color_continuous_scale="Cividis"
)

fig10.update_traces(textposition="outside")

fig10.update_layout(
    xaxis_tickangle=-45
)
```
